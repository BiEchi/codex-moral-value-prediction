```python
import pandas as pd
```

```python
# after train test split, we want to check the distribution of labels in each set
data = pd.read_json('../../../data/moral/interim/train_test/reddit/valid_bert.json', lines=True)
data
```

```python
print("num of non-moral tweets: ", len(data[data['Non-Moral'] == 1]))
print("num of moral tweets: ", len(data[data['Non-Moral'] == 0]))
print("ratio: ", len(data[data['Non-Moral'] == 0])/len(data[data['Non-Moral'] == 1]))
```

```python
label_list = data.columns.to_list()[2:]
label_num_dict = {label: 0 for label in label_list}

for label in label_list:
    label_num_dict[label] = data[label].sum()
```

```python
# draw a distribution of labels in data
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.figure(figsize=(10, 5), dpi=200)
plt.bar(label_num_dict.keys(), label_num_dict.values())
plt.title('Distribution of labels in data')
plt.xticks(rotation=45)
plt.show()
```
